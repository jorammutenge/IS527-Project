```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rcParams
from pyvis.network import Network
```

```python
# SET VIZ TEMPLATE
# colors
YELLOW='yellow'
BLACK='#1a1a1a'
GREY='#696969'
WHITE='white'

# fonts
futura_font_path = 'graphfonts/Futura/futura-medium-condensed.ttf'
FUTURA = fm.FontProperties(fname=futura_font_path)

paypal_font_path = 'graphfonts/Paypal/PayPalSansSmall-Regular.ttf'
PAYPAL = fm.FontProperties(fname=paypal_font_path)

DPI=200
```

```python
sheets = ['episode1', 'episode2', 'episode3', 'episode4', 'episode5', 'episode6']
for i, sheet in enumerate(sheets):
    globals()['df' + str(i + 1)] = pd.read_excel('data.xlsx', sheet_name = sheet)
```

```python
# func to clean df
def clean_df(data):
    return (data
 .assign(sorted_pairs=lambda df_: df_.apply(lambda df_: '-'.join(sorted([df_['initiator'], df_['responder']])), axis=1))
 .assign(weight=lambda df_: df_.groupby(['sorted_pairs'])['sorted_pairs'].transform('count'))
 .assign(dialogue=lambda df_: df_.groupby(['sorted_pairs'])['dialogue'].transform(lambda df_: ' '.join(df_.unique())))
 .drop_duplicates('sorted_pairs')
 .drop(columns='sorted_pairs')
)
```

```python
# clean dataframes
dataframes = [df1, df2, df3, df4, df5, df6]

for i in range(len(dataframes)):
    dataframes[i] = clean_df(dataframes[i])

df1, df2, df3, df4, df5, df6 = dataframes
```

```python
# func to create network
def create_graph(df):
    G = nx.from_pandas_edgelist(df, source='initiator', target='responder', edge_attr=['dialogue', 'weight'], create_using=nx.Graph())
    return G
```

```python
# create networks
graphs = []
for df in dataframes:
    graph = create_graph(df)
    graphs.append(graph)

G1, G2, G3, G4, G5, G6 = graphs
```

```python
# func to plot evolution of characters
def plot_evolution(df, ylabel):
    fig, ax = plt.subplots(figsize=(12,5), dpi=DPI, facecolor=BLACK)
    ax.set_facecolor(BLACK)
    for col in df.columns:
        ax.plot(df.index, df[col], label=col)

        # Add custom annotation
        if col == 'NICOLE':
            ax.annotate('NICOLE', xy=(2, df[col][2]), xytext=(.2, .46), fontsize=10, color=WHITE)
        elif col == 'TANYA':
            ax.annotate('TANYA', xy=(2, df[col][2]), xytext=(1.4, .38), fontsize=10, color=WHITE)
        elif col == 'RACHEL':
            ax.annotate('RACHEL', xy=(2, df[col][2]), xytext=(3.3, .21), fontsize=10, color=WHITE)
        elif col == 'SHANE':
            ax.annotate('SHANE', xy=(2, df[col][2]), xytext=(1.4, .21), fontsize=10, color=WHITE)
        elif col == 'ARMOND':
            ax.annotate('ARMOND', xy=(2, df[col][2]), xytext=(2.1, .65), fontsize=10, color=WHITE)

    # Customize the plot
    ax.tick_params(axis='both', colors=YELLOW, labelsize=10)
    ax.set_xticks(df.index)
    ax.set_xticklabels(['episode 1', 'episode 2', 'episode 3', 'episode 4', 'episode 5', 'episode 6'])
    ax.set_ylabel(ylabel, fontproperties=PAYPAL, color=WHITE, fontsize=12)
    ax.set_title('THE WHITE LOTUS\nseason one', fontproperties=FUTURA, color=YELLOW, fontsize=13)
    fig.suptitle('Evolution of characters', fontproperties=FUTURA, color=WHITE, fontsize=24, fontweight='bold', y=.96)

    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontproperties(PAYPAL)
        label.set_fontsize(12)
        label.set_color(YELLOW)

    for side in 'bottom,left'.split(','):
        ax.spines[side].set_edgecolor(WHITE)

    for side in 'top,right'.split(','):
        ax.spines[side].set_visible(False)

    # Add line labels
    lines = ax.get_lines()
    for i, line in enumerate(lines):
        line.set_label(df.columns[i])

    # ax.legend(frameon=False)
    plt.tight_layout()
    plt.show()
```

```python
# DEGREE CENTRALITY
# create list of degree centrality for all episodes
degree_evol = [nx.degree_centrality(graph) for graph in graphs]

# create dataframe from list
degree_evol_df = pd.DataFrame.from_records(degree_evol)

# set top five characters based on degree
top_5_degree = degree_evol_df[['NICOLE', 'TANYA', 'RACHEL', 'SHANE', 'ARMOND']]

# plot evolution of degree importance
plot_evolution(top_5_degree, 'Degree importance')
```
